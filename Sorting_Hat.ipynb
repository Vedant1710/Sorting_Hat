# Using GIF in Tkinter

```python
import tkinter as tk
from PIL import Image

root = tk.Tk()
file="Sorting_Hat.gif"

info = Image.open(file)

frames = info.n_frames  # gives total number of frames that gif contains

# creating list of PhotoImage objects for each frames
imgList = [tk.PhotoImage(file=file,format=f"gif -index {i}") for i in range(frames)]

count = 0
anim = None
# Function in which list iteration takes place, hece showing the gif, img by img
def animation(count):
    global anim
    im2 = imgList[count]

    gif_label.configure(image=im2)
    count += 1
    if count == frames:
        count = 0
    anim = root.after(50, lambda :animation(count))

def stop_animation():
    root.after_cancel(anim)

gif_label = tk.Label(root, image="") # A lable where ing will appear, initially no img would be presented
gif_label.pack()

start = tk.Button(root,text="start", command=lambda :animation(count))
start.pack()

stop = tk.Button(root,text="stop", command=stop_animation)
stop.pack()

root.mainloop()
```
